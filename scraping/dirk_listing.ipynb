```python
import requests
from bs4 import BeautifulSoup
url_category = 'https://www.dirk.nl/boodschappen'
response_category = requests.get(url_category)
soup_category = BeautifulSoup(response_category.text, 'html.parser')
categories_links = soup_category.find_all('a', class_='department')
links = [category['href'] for category in categories_links]
print(links)
```

```python
last_words = [link.split('/')[-1] for link in links]
print(last_words)
# Create a dictionary with last_words as keys and links as values
last_word_link_dict = {last_word: link for last_word, link in zip(last_words, links)}
print(last_word_link_dict)
```

```python
def scrape_sub_categories(url):
    response = requests.get(url)
    response.raise_for_status()
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    sub_categories = {}
    # Find all the sub-categories by their div class
    links = soup.find('div', class_='right')
    #print(links)
    link_tags = links.find_all('a')
    for links in link_tags:
        print(links)
        #sub_category_name = links.find('font')
        sub_category_name = links.text
        sub_category_link = links['href']
        sub_categories[sub_category_name] = "https://www.dirk.nl" + sub_category_link

    return sub_categories

# Initialize an empty dictionary to hold all subcategory data
all_sub_categories = {}

# Iterate over the main categories and collect sub-categories into a flat structure

for category, url in last_word_link_dict.items():
    sub_categories = scrape_sub_categories("https://www.dirk.nl"+url)
    all_sub_categories.update(sub_categories)
```

```python

from bs4 import BeautifulSoup
import re
def price_extract(price_div):
    print("entering price_extract")
    original_price = None
    offer_price = None

    if price_div:
        # Extract offer price
        offer_container = price_div.find('div', class_='price-container offer middle-item price')
        if offer_container:
            offer_price_div = offer_container.find('div', class_='price')
            if offer_price_div:
                offer_large = offer_price_div.find('span', class_='price-large')
                offer_small = offer_price_div.find('span', class_='price-small')
                if offer_large and offer_small:
                    offer_price = f"{offer_large.text.strip()}.{offer_small.text.strip()}"
                elif offer_large:
                    offer_price = f"0.{offer_large.text.strip()}"

        # Extract original price
            original_container = price_div.find('div', class_='price-container offer middle-item price')
            if original_container:
                original_label = original_container.find('div', class_='label price-label')
                if original_label:
                    original_price_span = original_label.find('span', class_='regular-price')
                    #print(original_price_span)
                    if original_price_span:
                        original_price = original_price_span.text.strip()
                        #print(original_price)

        else:
            #print("else")
            original_container=price_div.find('div', class_='price-container middle-item price')
            if original_container:
                original_price_div = original_container.find('div', class_='price')
                if original_price_div:
                    original_large = original_price_div.find('span', class_='hasEuros price-large')
                    
                    if(original_large):

                        original_small = original_price_div.find('span', class_='price-small')
                        original_price = f"{original_large.text.strip()}.{original_small.text.strip()}"
                        
                    else:
                        original_large_=original_price_div.find('span', class_='price-large')
                        original_price = f"0.{original_large_.text.strip()}"
                    
            offer_price = original_price
    return offer_price, original_price
```

```python
def extract_quantity_name_link(article):
    print("entering extract_quantity_name_link")
    #q=article.find('div', class_='bottom')
    name=article.find('a', class_='title').text if q else 'N/A'
    link = article.find('a')['href'] if q else 'N/A'

    quantity = article.find('span',class_='subtitle').text if q else 'N/A'
    return name, quantity,link
```
